```python
import csv
with open ('C:/Users/Tanisha/Desktop/2nd YEAR/RUGVED/PYTHON/bollywood-1.csv','r')as csvfile:
    reader= csv.DictReader(csvfile)
    data=[]
    for row in reader:
        data.append(row)
        
df=pd.DataFrame(data)
df.head()

FS_data = [row for row in data if row["ReleaseTime"] =="FS"]
df_FS_data = pd.DataFrame(FS_data)
df_FS_data.head()
```

```python
df['Year'] = df.apply(lambda row: row['Release Date'].split("-")[2],axis=1)

df['Year'].value_counts()
```
